```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
def job_finder():
    key = input('Enter All Your skills or designation: ').replace(' ','%20')
    df_dic = {}
    company_names = []
    skills_pc = []
    age_pc = []
    location_pc = []
    req_expirience_pc = []
    for seq in range(1,10):
        
        html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords='+key+'&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords='+key+'&pDate=I&sequence='+str(seq)+'&startPage=1').text
        soup = BeautifulSoup(html_text, 'lxml')
        jobs = soup.find_all('li',class_='clearfix job-bx wht-shd-bx')

        # print(job)
        for job in jobs:
            company_name = job.find('h3', class_='joblist-comp-name').text.replace(' ','').replace('\r','').replace('\n','')
            company_names.append(company_name)
            # print(company_name)
            skills =job.find('span',class_='srp-skills').text.replace(' ','').replace('\n','').replace('\r','')
            location =job.find('ul',class_='top-jd-dtl clearfix').span.text.replace(' ','').replace('\n','').replace('\r','')
            exp =job.find('ul',class_='top-jd-dtl clearfix').li.text.replace(' ','').replace('\n','').replace('\r','').replace('card_travel','')
#             print(location)
            location_pc.append(location)
            req_expirience_pc.append(exp)
            skills_pc.append(skills)
            # print(skills)
            age = job.find('span',class_='sim-posted').span.text.replace('\n','')
            age_pc.append(age)
        #     print(age)
    #         print(f"Company Name:\n{company_name}\nSkills: \n{skills}\nage:\n{age}\n---------------\n")


    df_dic['company_name'] = company_names
    df_dic['skills'] = skills_pc
    df_dic['age'] = age_pc
    df_dic['location'] = location_pc
    df_dic['req_expirience'] = req_expirience_pc
    # print(df_dic)  

    df = pd.DataFrame(df_dic)
    df.to_csv('file1.csv')
    return df
```
